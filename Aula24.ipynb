```python
xml_data = """
<data>
    <user>
        <name>joao</name>
        <email>joao@gmail.com</email>
        <cidade>santo andre</cidade>
    </user>
    <user>
        <name>maria</name>
        <email>maria@gmail.com</email>
        <!-- Nao existe -->
    </user>
</data>
"""

root = ET.fromstring(xml_data)
users = []

for user in root.findall("user"):
    users.append({
        'name':user.find('name').text if user.find('name') is not None else None,
        'email':user.find('email').text if user.find('email') is not None else None,
        'cidade':user.find('cidade').text if user.find('cidade') is not None else None

    })

df_xml = pd.DataFrame(users)
df_xml
```

```python
with open('data.json', 'r', encoding='utf-8') as f:
    data_arq = json.load(f)

data_arq[1]
```

```python
print('chaves', data_arq[0].keys())
```

```python
df_yelp = pd.json_normalize(data_arq)
df_yelp.head()
```

```python
df_yelpcustom = pd.DataFrame([{
    'business_id': d.get('business_id'),
    'name': d.get('name'),
    'city': d.get('city'),
    'review_count': d.get('review_count'),
    'stars': d.get('stars'),
    'categories': d.get('categories')
}for d in data_arq])

print(df_yelpcustom.info())
```

```python
df_yelpcustom.fillna('Desconhecido', inplace=True)
print(df_yelpcustom.info())
```

```python
df_yelpatri = pd.DataFrame([
    {
        'business_id': d.get('business_id'),
        'Wifi': ( d.get('attributes') or {} ).get('Wifi', 'Unknown'),
        'OutdoorSeating': ( d.get('attributes') or {} ).get('OutdoorSeating', 'Unknown'),
    } for d in data_arq])

print(df_yelpatri)
```

```python
data_arq[0].keys()
```

```python
weekdays = ['Sunday','Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']

df_horario = pd.DataFrame([
    {
        'business_id': d.get('business_id'),
        **{day: (d.get('hours') or {}).get(day, 'Fechado')   for day in weekdays}
}for d in data_arq])

print(df_horario.head())
```

```python
#df_yelpatri df_yelpcustom df_horario

df_merge = df_yelpcustom.merge(df_yelpatri, on='business_id', how='left')
df_merge = df_merge.merge(df_horario, on='business_id', how='left')

df_merge.info()
```

```python
df_merge_explode = df_merge.assign(categories=df_merge["categories"].str.split(', '))
```

```python
df_merge_explode = df_merge_explode.explode('categories')
df_merge_explode
```

```python
df_restaurants = df_merge_explode[df_merge_explode['categories'] == 'Restaurants']
len(df_restaurants)
```
