Explicação:
- `numpy` (`np`): biblioteca fundamental para manipulação de arrays e geração de dados simulados.
- `sklearn.linear_model.LinearRegression`: modelo de regressão linear simples da biblioteca Scikit-learn.
- `matplotlib.pyplot` (`plt`): biblioteca usada para criar gráficos, como o de dispersão (scatter plot).

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

Explicação:
- `np.random.seed(42)`: garante que os resultados da aleatoriedade sejam reproduzíveis.
- `X`: representa a variável independente (número de quartos), com 100 amostras aleatórias entre 0 e 2.
- `y`: variável dependente (preço da casa), definida por uma relação linear `y = 4 + 3X` com um ruído aleatório adicionado.

```python
np.random.seed(42)
X = 2 * np.random.rand(100, 1) # numero de quartos 0 a 2
y = 4 + 3 * X + np.random.rand(100, 1) # preço da casa com ruido

print('\nPrimeiros 5 exemplos:')
print(f'X (Quartos): {X[:5].flatten()}')
print(f'y (Preço): {y[:5].flatten()}')

plt.scatter(X, y, alpha=0.7)
plt.xlabel('Numero de quartos')
plt.ylabel('Preço da casa (x100k)')
plt.title('Relaçao de Quartos vs Preço')
plt.show()
```

Explicação:
- `LinearRegression()`: inicializa o modelo de regressão linear.
- `model.fit(X, y)`: treina o modelo com os dados sintéticos `X` e `y`.
- `model.intercept_`: valor do intercepto (b), ou seja, onde a reta cruza o eixo y.
- `model.coef_`: coeficiente angular (a), ou seja, a inclinação da reta (quanto o preço da casa aumenta por quarto adicional).

```python
model = LinearRegression()
model.fit(X, y)
print(f"\nIntercepto (b): {model.intercept_[0]:.2f}")           
print(f"\nCoeficiente angular (a): {model.coef_[0][0]:.2f}")    
```

Explicação:
- Aqui, fazemos uma previsão para uma nova entrada: uma casa com 1.5 quartos.
- O modelo retorna o valor estimado do preço da casa baseado na regressão treinada.

```python
X_new = np.array([[1.5]])   # Uma casa com 1.5 quartos
y_pred = model.predict(X_new)
print(f'Previsão para casa com 1.5 quartos: {y_pred[0][0]:.2f}x100k')

plt.scatter(X, y, alpha=0.7)
plt.plot(X, model.predict(X), color='red', linewidth=1)
plt.xlabel('Número de quartos')
plt.ylabel('Preço da casa (x100k)')
plt.title('Regressão Linear')
plt.show()
```

Agrupamento (clustering)

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# gerando dados sinteticos
X, y_true = make_blobs(
    n_samples = 300,
    centers = 4,
    cluster_std = 0.60,
    random_state = 0
)

print(f'\n Formato dos dados: {X.shape}')
plt.scatter(X[:,0], X[:, 1], c=y_true, s=50, cmap='viridis', alpha=0.7)
plt.title('Dados originais com clusters verdadeiros')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
y_kmeans = kmeans.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis', alpha=0.7)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title('Agrupamento cm k-means')
plt.show()
```

```python
print('\nCentroide de clusters:;')
for i, center in enumerate(centers):
    print(f'Cluester {i}: ({center[0]:.2f})')
```
