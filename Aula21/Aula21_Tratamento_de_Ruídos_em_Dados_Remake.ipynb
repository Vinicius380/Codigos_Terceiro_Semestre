# 2. Definição de Ruídos em Dados

Em análise de dados, o termo "ruído" refere-se a informações irrelevantes ou distorcidas que não contribuem para a análise ou aprendizado do modelo. Esse ruído pode ser introduzido durante a coleta de dados, erros de medição, falhas de sensores, ou até mesmo variações naturais nos dados que não são relevantes para a questão ou problema em questão.

Os ruídos podem dificultar a construção de modelos preditivos precisos, afetando a qualidade e a precisão das previsões. Portanto, é importante identificar e tratar esses ruídos para melhorar a qualidade dos modelos de machine learning.

Exemplos de ruídos incluem:

Erros de digitação nos dados.
Valores atípicos (outliers) que não representam o padrão dos dados.
Falhas no processo de coleta (como leituras de sensores com falhas).
Informações irrelevantes que não têm impacto no resultado da análise.
Em um processo de análise estatística e machine learning, a identificação e o tratamento de ruídos ajudam a melhorar a precisão e a robustez dos modelos.

## Exercício: Detectando outliers usando IQR

Definição:
A função usa o método do intervalo interquartil (IQR) para detectar valores extremos.

Exploração:
O IQR ajuda a separar os dados normais dos outliers estatísticos, facilitando a análise de anomalias.

Padronização:
Ao identificar outliers, pode-se decidir entre removê-los ou substituí-los por valores mais adequados.

Normalização:
Se os dados forem usados para treinamento de modelos, normalizar pode reduzir o impacto de valores extremos nos cálculos.



Implemente uma função para identificar outliers com o método do intervalo interquartil (IQR).

```python
# Define uma função para detectar outliers usando o método do Intervalo Interquartil (IQR)
def detectar_outliers_iqr(dados):  # A função recebe um conjunto de dados como entrada.
    Q1 = np.percentile(dados, 25)  # Calcula o primeiro quartil (Q1), que é o valor abaixo do qual 25% dos dados estão.
    Q3 = np.percentile(dados, 75)  # Calcula o terceiro quartil (Q3), que é o valor abaixo do qual 75% dos dados estão.
    IQR = Q3 - Q1  # Calcula o Intervalo Interquartil (IQR), que é a diferença entre Q3 e Q1.
    lim_inf = Q1 - 1.5 * IQR  # Calcula o limite inferior para identificar outliers.
    lim_sup = Q3 + 1.5 * IQR  # Calcula o limite superior para identificar outliers.
    return [x for x in dados if x < lim_inf or x > lim_sup]  # Retorna uma lista de valores que estão abaixo do limite inferior ou acima do limite superior.

# Aplica a função para detectar outliers no conjunto de dados 'dados_com_ruido'
outliers = detectar_outliers_iqr(dados_com_ruido)  # Chama a função passando o conjunto de dados 'dados_com_ruido'.
                                                  # A função retorna os outliers encontrados.

# Exibe os outliers identificados
print("Outliers identificados:", outliers)  # Imprime os outliers no console.
```

## Exercício: Detectando outliers com Z-score

Definição:
A função identifica outliers usando o método do Z-score, que mede a distância de cada ponto em relação à média.

Exploração:
O Z-score permite quantificar a intensidade de um outlier em relação à distribuição geral.

Padronização:
Os outliers podem ser removidos ou substituídos dependendo da necessidade da análise.

Normalização:
Se os dados forem usados em machine learning, normalizá-los pode reduzir a influência de valores extremos.




```python
from scipy.stats import zscore  # Importa a função 'zscore' da biblioteca SciPy, que calcula o escore z (z-score) para os dados.

# Define uma função para detectar outliers usando o método do z-score
def detectar_outliers_zscore(dados, limiar=3):  # A função recebe dois parâmetros:
                                               # - 'dados': uma lista ou array de valores numéricos.
                                               # - 'limiar': um valor limite para identificar outliers (padrão é 3).
    z_scores = np.abs(zscore(dados))  # Calcula o valor absoluto dos z-scores para cada ponto nos dados.
                                     # O z-score mede quantos desvios padrão um ponto está da média.
    return dados[z_scores > limiar]  # Retorna os valores que têm z-score maior que o limiar, considerados outliers.

# Aplica a função para detectar outliers em uma coluna específica do DataFrame
outliers_z = detectar_outliers_zscore(df['Valores'])  # Chama a função passando a coluna 'Valores' do DataFrame 'df'.
                                                     # A função retorna os outliers encontrados.

# Exibe os outliers detectados
print("Outliers pelo Z-score:", outliers_z.tolist())  # Converte os outliers para uma lista e imprime no console.
```

## Prática: Aplicando Normalização Min-Max

Definição:
A normalização transforma os dados para um intervalo [0,1].

Exploração:
O histograma permite visualizar a distribuição dos dados normalizados.

Padronização:
Garante que os valores estejam em uma escala uniforme.

Normalização:
Facilita a convergência de modelos de aprendizado de máquina ao remover diferenças de escala.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['Normalizado'] = scaler.fit_transform(df[['Valores']])

# Histograma dos dados normalizados
plt.hist(df['Normalizado'], bins=20, color='seagreen', edgecolor='black')
plt.title("Dados Normalizados")
plt.xlabel("Valores")
plt.ylabel("Frequência")
plt.show()
```

# Exercício: Criando uma função de normalização


Crie uma função para normalizar qualquer conjunto de dados entre um intervalo arbitrário.

```python
def normalizar_dados(dados, min_val=0, max_val=1):
    return (dados - min(dados)) / (max(dados) - min(dados)) * (max_val - min_val) + min_val

dados_normalizados = normalizar_dados(df['Valores'])
print(dados_normalizados)
```
